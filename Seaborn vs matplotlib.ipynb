# Matplotlib

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
#Cоздаем DF из выживших и их возраста
Sur = df[['Age', 'Survived']].query('Survived == 1')
````

````python
#Cоздаем DF из погибших и их возраста
Dec = df[['Age', 'Survived']].query('Survived == 0')
````

````python
#выжившие по возрасту
Sur['Age'].plot.hist(edgecolor = 'k')
plt.xlabel('age_range')
plt.ylabel('count')
plt.title('Survived')
````

````python
#не выжили по возрасту
Dec['Age'].plot.hist(edgecolor = 'k')
plt.xlabel('age_range')
plt.ylabel('count')
plt.title('Deceased')
````

# Seaborn

````python
import seaborn as sns
````

````python
#выжившие и погибшие по возрасту
sns.displot(data=df, x="Age", col='Survived')
````

````python
#выжившие и погибшие по классу круиза
sns.boxplot(x='Survived', y="Pclass", data=df)
````

````python
#группировка по выжившим и погибшим исходя из их класса круиза и возраста
sns.boxplot(x='Pclass', y='Age', hue='Survived', data=df)
````
